```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Impactful batters in last 3 seasons of ipl

```python
t20_2023 = pd.read_csv('t20_2023.csv')
t20_2024 = pd.read_csv('t20_2024.csv')
t20_2022 = pd.read_csv('t20_2022.csv')
```

```python
t20_merged = pd.concat([t20_2022, t20_2023, t20_2024],ignore_index = True)
```

```python
t20_merged.head()
```

```python
ipl_data = t20_merged[t20_merged['competition'].str.lower() == 'ipl']
```

```python
ipl_data.head()
```

```python
pd.set_option('display.max_columns', None)
print(ipl_data.iloc[0].to_frame().T)
```

```python
data = ipl_data[['bat', 'batruns','ballfaced','dismissal', 'p_match']]
data['batruns'] = pd.to_numeric(data['batruns'], errors='coerce')

bat_stats = data.groupby('bat').agg(
    matches_played=('p_match', 'nunique'),
    total_runs=('batruns', 'sum'),
    balls_faced=('ballfaced', 'sum'),
    dismissals=('dismissal', 'count'),
).reset_index()

bat_stats['Average'] = round((bat_stats['total_runs']/bat_stats['dismissals']),2)
bat_stats['Strike Rate'] = round((bat_stats['total_runs'] / bat_stats['balls_faced']) * 100, 2)

eligible_batters = bat_stats[
    (bat_stats['matches_played'] > 25) & (bat_stats['total_runs'] > 500)
]['bat']

filtered_data = data[data['bat'].isin(eligible_batters)]

results = []

for batter, group in filtered_data.groupby('bat'):
    group = group.reset_index(drop=True)
    consecutive_streaks = 0
    streak_count = 0
    
    for idx, row in group.iterrows():
        if row['batruns'] in [0, 1]:
            streak_count += 1
        else:
            if streak_count > 0:
                consecutive_streaks += streak_count
                streak_count = 0
    if streak_count > 0:  # Add remaining streaks
        consecutive_streaks += streak_count

    total_balls = len(group)
    avg_dot_single_streak_percent = (consecutive_streaks / total_balls) * 100 if total_balls > 0 else 0

    boundary_indices = group[group['batruns'] >= 4].index
    gaps = boundary_indices.to_series().diff().dropna()
    avg_gap = gaps.mean() if not gaps.empty else 0

    results.append({
        'bat': batter,
        'Mean Consecutive Dots/Singles_percent': avg_dot_single_streak_percent,
        'avg_gap_between_boundaries': avg_gap
    })

summary_df = pd.DataFrame(results)
summary_df = summary_df.merge(bat_stats, on='bat')

# Scaling avg_dot_single_streak_percent to 1-10 range
min_streak = summary_df['Mean Consecutive Dots/Singles_percent'].min()
max_streak = summary_df['Mean Consecutive Dots/Singles_percent'].max()

summary_df['Mean Consecutive Dots/Singles_scaled'] = (
    1 + ((summary_df['Mean Consecutive Dots/Singles_percent'] - min_streak) / (max_streak - min_streak)) * 9
)

summary_df[['bat', 'Mean Consecutive Dots/Singles_scaled', 'avg_gap_between_boundaries', 'total_runs', 'Average', 'Strike Rate']].sort_values(by='total_runs', ascending=False)
```

```python
from scipy.stats import zscore
```

```python
summary_df['Inverted Mean Consecutive Dots/Singles'] = 100 - summary_df['Mean Consecutive Dots/Singles_percent']
```

```python
summary_df['Inverted Mean avg_gap_between_boundaries'] = 100 - summary_df['avg_gap_between_boundaries']
```

```python
features = ['Inverted Mean Consecutive Dots/Singles', 'Inverted Mean avg_gap_between_boundaries', 'Average', 'Strike Rate']
```

```python
summary_df[features] = summary_df[features].apply(zscore)
```

```python
summary_df['Impact Score'] = summary_df[features].mean(axis=1)
```

```python
min_impact = summary_df['Impact Score'].min()
max_impact = summary_df['Impact Score'].max()
```

```python
summary_df['Impact Score (1-10)'] = round((1 + ((summary_df['Impact Score'] - min_impact) / (max_impact - min_impact)) * 9),2)
```

```python
summary_df[['bat', 'Impact Score (1-10)']].sort_values(by='Impact Score (1-10)', ascending=False).reset_index(drop=True)
```
